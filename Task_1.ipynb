```python
import pandas as pd
file_path = '/content/drive/MyDrive/assessment/Task 1.xlsx'
excel_data = pd.ExcelFile(file_path)
excel_data.sheet_names
```

```python
df = excel_data.parse('Task')
df1 = excel_data.parse('Taxonomy')
display(df.head())
display(df1.head())
```

```python
root_cause_dict = dict(zip(df1['Root Cause'], df1['Root Cause']))
symptom_condition_dict = dict(zip(df1['Symptom Condition '], df1['Symptom Condition ']))
symptom_component_dict = dict(zip(df1['Symptom Component'], df1['Symptom Component']))
fix_condition_dict = dict(zip(df1['Fix Condition'], df1['Fix Condition']))
fix_component_dict = dict(zip(df1['Fix Component'], df1['Fix Component']))
```

```python

def tag_data(complaint, cause, correction):
    root_cause = []
    symptom_conditions = []
    symptom_components = []
    fix_conditions = []
    fix_components = []

    complaint = str(complaint).lower() if not pd.isna(complaint) else ''
    cause = str(cause).lower() if not pd.isna(cause) else ''
    correction = str(correction).lower() if not pd.isna(correction) else ''

    for key in root_cause_dict:
        if str(key).lower() in complaint or str(key).lower() in cause or str(key).lower() in correction:
            root_cause.append(key)
            break

    for key in symptom_condition_dict:
        if str(key).lower() in complaint or str(key).lower() in cause:
            symptom_conditions.append(key)
            if len(symptom_conditions) == 3:
              break

    for key in symptom_component_dict:
      if str(key).lower() in complaint or str(key).lower() in cause:
          symptom_components.append(key)
          if len(symptom_components) == 3:
            break

    for key in fix_condition_dict:
        if str(key).lower() in correction:
            fix_conditions.append(key)
            if len(fix_conditions) == 3:
                break

    for key in fix_component_dict:
        if str(key).lower() in correction:
            fix_components.append(key)
            if len(fix_components) == 3:
              break

    return (
        root_cause[0] if root_cause else '',
        symptom_conditions[:3] + [''] * (3 - len(symptom_conditions[:3])),
        symptom_components[:3] + [''] * (3 - len(symptom_components[:3])),
        fix_conditions[:3] + [''] * (3 - len(fix_conditions[:3])),
        fix_components[:3] + [''] * (3 - len(fix_components[:3]))
    )


for index, row in df.iterrows():
    results = tag_data(row['Complaint'], row['Cause'], row['Correction'])
    df.loc[index, 'Root Cause'] = results[0]
    df.loc[index, ['Symptom Condition 1', 'Symptom Condition 2', 'Symptom Condition 3']] = results[1]
    df.loc[index, ['Symptom Component 1', 'Symptom Component 2', 'Symptom Component 3']] = results[2]
    df.loc[index, ['Fix Condition 1', 'Fix Condition 2', 'Fix Condition 3']] = results[3]
    df.loc[index, ['Fix Component 1', 'Fix Component 2', 'Fix Component 3']] = results[4]

df.to_csv('tagged_data.csv', index=False)
df
```
